```python
def count(n):
    return n+1

def MinMax_Search(game, state):
    global c
    c = 1
    player = game.To_Move(state)
    if player=='X':
        v, move = Max_Value(game,state)
    else:
        v, move = Min_Value(game,state)
    return move

def Max_Value(game,state):
    global c
    c = count(c)
    player = game.To_Move(state)
    v=float("-inf")
    if game.IsTerminal(state):
        return game.Utility(state,player), None
    
    for a in game.Actions(state):
        v2,a2 =  Min_Value(game,game.Result(state,a))
        if v2>v:
            v,move = v2,a
    return v, move

def Min_Value(game,state):
    global c
    c = count(c)
    
    player = game.To_Move(state)
    v=float("inf")
    
    if game.IsTerminal(state):
        return game.Utility(state,player), None
    for a in game.Actions(state):
        v2,a2 = Max_Value(game,game.Result(state,a))
        if v2<v:
            v,move = v2,a
    return v, move

def Alpha_Beta_Search(game, state):
    global c_alpha 
    c_alpha =1
    player = game.To_Move(state)
    if player=='X':
        v, move = Max_Value_alpha(game,state,float("-inf"),float("inf"))
    else:
        v, move = Min_Value_alpha(game,state,float("-inf"),float("inf"))
    return move

def Max_Value_alpha(game,state,alpha,beta):
    global c_alpha
    c_alpha = count(c_alpha)
    player = game.To_Move(state)
    v=float("-inf")
    if game.IsTerminal(state):
        return game.Utility(state,player), None
    
    for a in game.Actions(state):
        v2,a2 =  Min_Value_alpha(game,game.Result(state,a),alpha,beta)
        if v2>v:
            v,move = v2,a
            alpha =max(alpha,v)
        if v>=beta:
            return v,move
    return v, move

def Min_Value_alpha(game,state,alpha,beta):
    global c_alpha
    c_alpha = count(c_alpha)
    player = game.To_Move(state)
    v=float("inf")    
    if game.IsTerminal(state):
        return game.Utility(state,player), None
    
    for a in game.Actions(state):
        v2,a2 = Max_Value_alpha(game,game.Result(state,a),alpha,beta)
        if v2<v:
            v,move = v2,a
            beta=min(beta,v)
        if v<=alpha:
            return v,move
    return v, move


class Tic_Tac_Toe:
    def __init__(self,first):
        self.board = [[' ' for _ in range(3) ] for _ in range(3)]
        self.first = first
        if self.first == 'X':
            self.second = 'O'
        elif self.first == 'O':
            self.second = 'X'

    
    def To_Move(self,state): # whose turn?
        total_first = sum(row.count(self.first) for row in state)
        total_second = sum(row.count(self.second) for row in state)
        if total_first>total_second:
            return self.second
        return self.first
        
        
    def IsTerminal(self,state): # check the end 
        
        # 1st possible case : row 
        for row in state:
            if row.count('X')==3 or row.count('O')==3:
                #print('type 1')
                return True
            
        # 2nd possible case : column
        for col in range(3):
            if (state[0][col]==state[1][col]==state[2][col]) and (state[0][col] == 'O' or state[0][col] == 'X'):
                #print('type 2')
                return True
        
        # 3rd possible case :crossline (top left to bottom right)
        if (state[0][0]==state[1][1]==state[2][2]) and (state[0][0] =='O' or state[0][0] =='X'):
            #print('type 3')
            return True
        
        # 4th possible case :crossline( top right to bottom left)
        if (state[2][0]==state[1][1]==state[0][2]) and (state[2][0] =='O' or state[2][0] =='X'):
            #print('type 4')
            return True
        
        # it means X row, column, crosslines.->two possible case : 1. empty blank 2. full 
        # 5th possible case: draw
        for row in state:
            if ' ' in row:
                #print('type 5')
                return False # it means there is at least one empty 
            
        #print('type 6')
        return True # it means draw
        
        
    def Utility(self,state,player):
        
        # check the end 
        if player=='X':
            non_player='O'
            #print('Utility : ',player)
        else:
            non_player='X'
        
        if self.IsTerminal(state): # end 
            #print('Utility is Terminal')
            # who wins?
            if player=='X':
                player='X'
                non_player='O'
                for row in state:
                    if row.count(player)==3:
                        return 1
                    if row.count(non_player)==3:
                        return -1
                
                for col in range(3):
                    if (state[0][col]==state[1][col]==state[2][col]) and (state[0][col] == player):
                        return 1
                    if (state[0][col]==state[1][col]==state[2][col]) and (state[0][col] == non_player):
                        return -1
                if (state[0][0]==state[1][1]==state[2][2]) and (state[0][0] ==player):
                    return 1
            
                if (state[0][0]==state[1][1]==state[2][2]) and (state[0][0] ==non_player):
                    return -1
            
                if (state[2][0]==state[1][1]==state[0][2]) and (state[2][0] ==player):
                    return 1
            
                if (state[2][0]==state[1][1]==state[0][2]) and (state[2][0] ==non_player):
                    return -1
            else: # player = 'O'
                player='O'
                non_player='X'
                for row in state:
                    if row.count(player)==3:
                        return -1
                    if row.count(non_player)==3:
                        return 1
                
                for col in range(3):
                    if (state[0][col]==state[1][col]==state[2][col]) and (state[0][col] == player):
                        return -1
                    if (state[0][col]==state[1][col]==state[2][col]) and (state[0][col] == non_player):
                        return 1
                if (state[0][0]==state[1][1]==state[2][2]) and (state[0][0] ==player):
                    return -1
            
                if (state[0][0]==state[1][1]==state[2][2]) and (state[0][0] ==non_player):
                    return 1
            
                if (state[2][0]==state[1][1]==state[0][2]) and (state[2][0] ==player):
                    return -1
            
                if (state[2][0]==state[1][1]==state[0][2]) and (state[2][0] ==non_player):
                    return 1
            return 0
        else:
            return 99

        
                    
    def Actions(self,state): #possible actions (find the empty)
        actions=[]
        for i in range(3):
            for j in range(3):
                if state[i][j]==' ':
                    actions.append((i,j)) 
        return actions
        
        
    def Result(self,state,action): #place 'X' or 'O'
        i,j = action
        player = self.To_Move(state)
        new_state = [row.copy() for row in state]
        new_state[i][j] = player
        return new_state

    def display(self, state):
        i=0
        for row in state:
            print("  l".join(row))
            if i<2:
                print('---+---+---')
            i+=1
    def transformer(self,state):
        actions = self.Actions(state)
        lst=[]
        for a in actions:
            if a==(0,0):
                lst.append(1)
            if a==(0,1):
                lst.append(2)
            if a==(0,2):
                lst.append(3)
            if a==(1,0):
                lst.append(4)
            if a==(1,1):
                lst.append(5)
            if a==(1,2):
                lst.append(6)
            if a==(2,0):
                lst.append(7)
            if a==(2,1):
                lst.append(8)
            if a==(2,2):
                lst.append(9)
        return lst[:]
    def twoD2oneD(self,action):
        if action ==(0,0):
            return 1
        if action ==(0,1):
            return 2
        if action ==(0,2):
            return 3
        if action ==(1,0):
            return 4
        if action ==(1,1):
            return 5
        if action ==(1,2):
            return 6
        if action ==(2,0):
            return 7
        if action ==(2,1):
            return 8
        if action ==(2,2):
            return 9
        
    
    def inverse_transformer(self,action):
        if action ==1:
            return (0,0)
        if action ==2:
            return (0,1)
        if action ==3:
            return (0,2)
        if action ==4:
            return (1,0)
        if action ==5:
            return (1,1)
        if action==6:
            return (1,2)
        if action==7:
            return (2,0)
        if action==8:
            return (2,1)
        if action==9:
            return (2,2)
        print('????????')
        
```

```python
def MinMax_Search(game, state):
    global n
    n=1
    player = game.To_Move(state)
    #print('MinMax_Search : ', player)
    #print(player,"'s turn")
    #print("original state : ")
    #for row in state:
    #    print(row)
    if player=='X':
        v, move = Max_Value(game,state)
    else:
        v, move = Min_Value(game,state)
    return move

def Max_Value(game,state):
    global n
    n= count(n)
    player = game.To_Move(state)
    #print('Max_Value_player : ', player)
    #print(player,"'s turn in Max_value function")
    v=float("-inf")
    if game.IsTerminal(state):
        #print("Before action, the game was already finished")
        #print("Because the state :  ")
        #for row in state:
        #    print(row)
        #print('game is Terminal in Max_Value')
        #print('Therefore, Utility is :',game.Utility(state,player))
        return game.Utility(state,player), None
    
    for a in game.Actions(state):
        #print(player,"'s action : ",a)
        #print('action : ',a)
        #for row in game.Result(state,a):
        #    print(row)
        v2,a2 =  Min_Value(game,game.Result(state,a))
        #print('v2, a2:',v2,a2)
        #print('In Max_value function, after',player,"'s action, the state : ")
        #for row in game.Result(state,a):
        #    print(row)
        if v2>v:
        #    print("current v : ",v,"current v2 :",v2)
        #    print('v2>v, so')
        #    print("v,move = v2, a")
            v,move = v2,a
        #    print('after changing, v, move = ',v,move)
        #print('------------')
    return v, move

def Min_Value(game,state):
    global n
    n= count(n)
    player = game.To_Move(state)
    #print('Min_value_player :',player)
    #print(player,"'s turn in Max_value function")
    v=float("inf")
    
    if game.IsTerminal(state):
        #print("Before action, the game was already finished")
        #print("Because the state :  ")
        #for row in state:
        #    print(row)
        #print('Therefore, Utility is :',game.Utility(state,player))
        return game.Utility(state,player), None
    

    #print(player,'is not Terminal in Min')
    for a in game.Actions(state):
       # print(player,"'s action : ",a)
        #for row in game.Result(state,a):
         #   print(row)
        #print('action : ',a)
        v2,a2 = Max_Value(game,game.Result(state,a))
       # print('In Max_value function, after',player,"'s action, the state : ")
        #for row in game.Result(state,a):
        #    print(row)
        #print('v2, a2:',v2,a2)
        if v2<v:
        #    print("current v : ",v,"current v2 :",v2)
        #    print('v2<v, so')
        #    print("v,move = v2, a")
            v,move = v2,a
        #    print('after changing, v, move = ',v,move)
        #print('-------------------')

    return v, move

def Alpha_Beta_Search(game, state):
    global c
    c=1
    player = game.To_Move(state)
    if player=='X':
        v, move = Max_Value_alpha(game,state,float("-inf"),float("inf"))
    else:
        v, move = Min_Value_alpha(game,state,float("-inf"),float("inf"))
    return move

def Max_Value_alpha(game,state,alpha,beta):
    player = game.To_Move(state)
    v=float("-inf")
    if game.IsTerminal(state):
        return game.Utility(state,player), None
    
    for a in game.Actions(state):
        v2,a2 =  Min_Value_alpha(game,game.Result(state,a),alpha,beta)
        if v2>v:
            v,move = v2,a
            alpha =max(alpha,v)
        if v>=beta:
            return v,move
    return v, move

def Min_Value_alpha(game,state,alpha,beta):
    player = game.To_Move(state)
    v=float("inf")    
    if game.IsTerminal(state):
        return game.Utility(state,player), None
    
    for a in game.Actions(state):
        v2,a2 = Max_Value_alpha(game,game.Result(state,a),alpha,beta)
        if v2<v:
            v,move = v2,a
            beta=min(beta,v)
        if v<=alpha:
            return v,move
    return v, move


class Tic_Tac_Toe:
    def __init__(self,first):
        self.board = [[' ' for _ in range(3) ] for _ in range(3)]
        self.first = first
        if self.first == 'X':
            self.second = 'O'
        elif self.first == 'O':
            self.second = 'X'

    
    def To_Move(self,state): # whose turn?
        total_first = sum(row.count(self.first) for row in state)
        total_second = sum(row.count(self.second) for row in state)
        if total_first>total_second:
            return self.second
        return self.first
        
        
    def IsTerminal(self,state): # check the end 
        
        # 1st possible case : row 
        for row in state:
            if row.count('X')==3 or row.count('O')==3:
                #print('type 1')
                return True
            
        # 2nd possible case : column
        for col in range(3):
            if (state[0][col]==state[1][col]==state[2][col]) and (state[0][col] == 'O' or state[0][col] == 'X'):
                #print('type 2')
                return True
        
        # 3rd possible case :crossline (top left to bottom right)
        if (state[0][0]==state[1][1]==state[2][2]) and (state[0][0] =='O' or state[0][0] =='X'):
            #print('type 3')
            return True
        
        # 4th possible case :crossline( top right to bottom left)
        if (state[2][0]==state[1][1]==state[0][2]) and (state[2][0] =='O' or state[2][0] =='X'):
            #print('type 4')
            return True
        
        # it means X row, column, crosslines.->two possible case : 1. empty blank 2. full 
        # 5th possible case: draw
        for row in state:
            if ' ' in row:
                #print('type 5')
                return False # it means there is at least one empty 
            
        #print('type 6')
        return True # it means draw
        
        
    def Utility(self,state,player):
        
        # check the end 
        if player=='X':
            non_player='O'
            #print('Utility : ',player)
        else:
            non_player='X'
        
        if self.IsTerminal(state): # end 
            #print('Utility is Terminal')
            # who wins?
            if player=='X':
                player='X'
                non_player='O'
                for row in state:
                    if row.count(player)==3:
                        return 1
                    if row.count(non_player)==3:
                        return -1
                
                for col in range(3):
                    if (state[0][col]==state[1][col]==state[2][col]) and (state[0][col] == player):
                        return 1
                    if (state[0][col]==state[1][col]==state[2][col]) and (state[0][col] == non_player):
                        return -1
                if (state[0][0]==state[1][1]==state[2][2]) and (state[0][0] ==player):
                    return 1
            
                if (state[0][0]==state[1][1]==state[2][2]) and (state[0][0] ==non_player):
                    return -1
            
                if (state[2][0]==state[1][1]==state[0][2]) and (state[2][0] ==player):
                    return 1
            
                if (state[2][0]==state[1][1]==state[0][2]) and (state[2][0] ==non_player):
                    return -1
            else: # player = 'O'
                player='O'
                non_player='X'
                for row in state:
                    if row.count(player)==3:
                        return -1
                    if row.count(non_player)==3:
                        return 1
                
                for col in range(3):
                    if (state[0][col]==state[1][col]==state[2][col]) and (state[0][col] == player):
                        return -1
                    if (state[0][col]==state[1][col]==state[2][col]) and (state[0][col] == non_player):
                        return 1
                if (state[0][0]==state[1][1]==state[2][2]) and (state[0][0] ==player):
                    return -1
            
                if (state[0][0]==state[1][1]==state[2][2]) and (state[0][0] ==non_player):
                    return 1
            
                if (state[2][0]==state[1][1]==state[0][2]) and (state[2][0] ==player):
                    return -1
            
                if (state[2][0]==state[1][1]==state[0][2]) and (state[2][0] ==non_player):
                    return 1
            return 0
        else:
            return 99

        
                    
    def Actions(self,state): #possible actions (find the empty)
        actions=[]
        for i in range(3):
            for j in range(3):
                if state[i][j]==' ':
                    actions.append((i,j)) 
        return actions
        
        
    def Result(self,state,action): #place 'X' or 'O'
        i,j = action
        player = self.To_Move(state)
        new_state = [row.copy() for row in state]
        new_state[i][j] = player
        return new_state

    def display(self, state):
        i=0
        for row in state:
            print("  l".join(row))
            if i<2:
                print('---+---+---')
            i+=1
    def transformer(self,state):
        actions = self.Actions(state)
        lst=[]
        for a in actions:
            if a==(0,0):
                lst.append(1)
            if a==(0,1):
                lst.append(2)
            if a==(0,2):
                lst.append(3)
            if a==(1,0):
                lst.append(4)
            if a==(1,1):
                lst.append(5)
            if a==(1,2):
                lst.append(6)
            if a==(2,0):
                lst.append(7)
            if a==(2,1):
                lst.append(8)
            if a==(2,2):
                lst.append(9)
        return lst[:]
    def twoD2oneD(self,action):
        if action ==(0,0):
            return 1
        if action ==(0,1):
            return 2
        if action ==(0,2):
            return 3
        if action ==(1,0):
            return 4
        if action ==(1,1):
            return 5
        if action ==(1,2):
            return 6
        if action ==(2,0):
            return 7
        if action ==(2,1):
            return 8
        if action ==(2,2):
            return 9
        
    
    def inverse_transformer(self,action):
        if action ==1:
            return (0,0)
        if action ==2:
            return (0,1)
        if action ==3:
            return (0,2)
        if action ==4:
            return (1,0)
        if action ==5:
            return (1,1)
        if action==6:
            return (1,2)
        if action==7:
            return (2,0)
        if action==8:
            return (2,1)
        if action==9:
            return (2,2)
        print('????????')
        
```
